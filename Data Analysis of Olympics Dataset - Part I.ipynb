There are several special characters used in the values of column Team such as United States-1, New York Athletic Club # 1, Denmark/Sweden. So the task is to remove these characters.

```python
pysqldf('''
        select distinct Team
        from athlete_events
        where Team like '%United%';
        ''')
```

```python
athlete_events["Team"] = athlete_events["Team"].replace("United States-3", 'United States').replace("United States-4", 'United States')
athlete_events["Team"] = athlete_events["Team"].replace("United States-5", 'United States').replace("United States-6", 'United States')
athlete_events["Team"] = athlete_events["Team"].replace("United States-7", 'United States').replace("United States-8", 'United States')
athlete_events["Team"] = athlete_events["Team"].replace("United States-9", 'United States').replace("United States-10", 'United States')
athlete_events["Team"] = athlete_events["Team"].replace("United States-11", 'United States').replace("United States-12", 'United States')
athlete_events["Team"] = athlete_events["Team"].replace("United States-13", 'United States').replace("United States-14", 'United States')
athlete_events["Team"] = athlete_events["Team"].replace("United States Virgin Islands-1", 'United States Virgin Islands').replace("United States Virgin Islands-2", 'United States Virgin Islands')
```

### 1.3 Initial Data Exploration

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### What is the total number of olympic medals won?

```python
pysqldf('''
        select count(Medal) as total_medals_won
        from athlete_events
        where Medal in ('Gold', 'Silver', 'Bronze');
        ''')
```

**Answer:** 39783

#### Top ten names of countries (NOC) with most olympic medals?

```python
topten_noc = pysqldf('''
        select NOC, count(Medal) as medal_count
        from athlete_events
        where Medal in ('Gold', 'Silver', 'Bronze')
        group by NOC
        order by medal_count desc
        limit 10;
        ''')

topten_noc[:]
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.barplot(x='NOC', y='medal_count', data=topten_noc, palette='magma')
```

#### Top ten sports with most olympic medals?

```python
topten_sprt = pysqldf('''
        select Sport, count(Medal) as medal_count_sport
        from athlete_events
        where Medal in ('Gold', 'Silver', 'Bronze')
        group by Sport
        order by medal_count_sport desc
        limit 10;
        ''')

topten_sprt[:]
```

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.barplot(x='medal_count_sport', y='Sport', data=topten_sprt, palette='magma', orient='h')
```
